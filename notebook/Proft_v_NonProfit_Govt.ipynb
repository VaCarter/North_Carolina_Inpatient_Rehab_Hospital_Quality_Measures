```python
import pandas as pd
```

```python
Facility_Info_25= pd.read_csv('../data/Inpatient_Rehabilitation_Facility-General_Information_Jun2025.csv')
Facility_Info_25.head()
```

```python
Provider_Data_25= pd.read_csv('../data/Inpatient_Rehabilitation_Facility_Provider_Data_Jun2025.csv')
Provider_Data_25
```

```python
Provider_Data_NC_25= Provider_Data_25.loc[Provider_Data_25['State']=='NC'].reset_index(drop=True)
Provider_Data_NC_25.head()
```

```python
Provider_Data_NC_25=Provider_Data_NC_25[['Provider Name','County/Parish','City/Town']]
Provider_Data_NC_25.head()
```

```python
Facility_Info_NC_25=Facility_Info_25.loc[Facility_Info_25['State']=='NC'].reset_index(drop=True)
Facility_Info_NC_25.head()
```

```python
Facility_Info_NC_25=Facility_Info_NC_25[['Provider Name','County/Parish','City/Town','Ownership Type']]
Facility_Info_NC_25
```

Merging facility and provider data 2025:

```python
Facility_Provider_NC_2025= pd.merge(Provider_Data_NC_25, Facility_Info_NC_25,
                                 on='Provider Name',
                                 how= 'left')
Facility_Provider_NC_2025.head()
```

```python
Facility_Provider_NC_2025=Facility_Provider_NC_2025.drop_duplicates().reset_index(drop=True)
Facility_Provider_NC_2025.head(3)
```

```python
Facility_Provider_NC_2025=Facility_Provider_NC_2025[['Provider Name','County/Parish_x','City/Town_x','Ownership Type']]
Facility_Provider_NC_2025
```

```python
Facility_Provider_NC_2025.info()
```

```python
Facility_Provider_NC_2025_grouped=Facility_Provider_NC_2025.groupby('Ownership Type')['County/Parish_x'].value_counts()
Facility_Provider_NC_2025_grouped
```

```python
Facility_Provider_NC_2025_grouped=Facility_Provider_NC_2025_grouped.to_frame().reset_index()
```

```python
Facility_Provider_NC_2025_grouped
```

```python
NC_IRF_Counties_List=['Buncombe','Catawba','Craven','Cumberland','Durham','Forsyth','Guilford','Lenoir','Mecklenburg','Moore','Nash','New_Hanover','Orange','Pitt','Rowan','Vance','Wake']
```

```python
Facility_Provider_NC_2025_grouped=Facility_Provider_NC_2025_grouped.loc[Facility_Provider_NC_2025_grouped['County/Parish_x'].isin(NC_IRF_Counties_List)].reset_index(drop=True)
Facility_Provider_NC_2025_grouped.head()
```

```python
Facility_Provider_NC_2025_grouped=Facility_Provider_NC_2025_grouped.rename(columns={'County/Parish_x':'County'})
Facility_Provider_NC_2025_grouped.head(3)
```

```python
Facility_Provider_NC_2025_grouped.to_csv('Facility_Provider_NC_2025_grouped.csv', index=False)
```

```python
region_grouped_25=pd.read_csv('../data/region_grouped_25.csv')
```

```python
region_grouped_25_filtered=region_grouped_25[['County','Region','Age 65 Plus']]
region_grouped_25_filtered
```

```python
region_grouped_25_filtered.isna().sum()
```

```python
region_grouped_25_filtered=region_grouped_25_filtered.dropna()
region_grouped_25_filtered
```

```python
Region_Ownership_Type=pd.merge(Facility_Provider_NC_2025_grouped, region_grouped_25_filtered,
                               on='County',
                               how='left')
Region_Ownership_Type
```

```python
Region_Ownership_Type=Region_Ownership_Type[['Ownership Type','County','Region']]
Region_Ownership_Type
```

```python
Region_Ownership_Type=Region_Ownership_Type.drop_duplicates().reset_index(drop=True)
Region_Ownership_Type
```

```python
Region_Ownership_Type.to_csv('Region_Ownership_Type.csv')
```
