## Filtering down to NC Counties from 23-25:

```python
IRF_provider_data_June_23 = pd.read_csv('../data/Inpatient_Rehabilitation_Facility-Provider_Data_Jun2023.csv')
IRF_provider_data_June_23.head(1)
```

```python
NC_Readmissions_2023=IRF_provider_data_June_23.loc[(IRF_provider_data_June_23['Measure Code'] == 'I_018_01_PPR_WI_RSRR') & (IRF_provider_data_June_23['State']=='NC')].reset_index(drop=True)
NC_Readmissions_2023.head()
```

```python
drop_rows= NC_Readmissions_2023.loc[NC_Readmissions_2023["Score"]== "Not Available"]
```

```python
NC_Readmissions_2023=NC_Readmissions_2023.drop(drop_rows.index, axis=0)
```

```python
NC_Readmissions_2023['Score']=NC_Readmissions_2023['Score'].astype(float).reset_index(drop=True)
```

```python
NC_Readmissions_2023=NC_Readmissions_2023[['CMS Certification Number (CCN)', 'County Name','State','Measure Code','Score']].rename(columns={'Score':'2023 Score','County Name':'County/Parish'})
```

```python
Avg_NC_Readmissions_2023=NC_Readmissions_2023.groupby('County/Parish')['2023 Score'].mean().to_frame()
```

```python
Avg_NC_Readmissions_2023=Avg_NC_Readmissions_2023.reset_index()
Avg_NC_Readmissions_2023
```

```python
IRF_provider_data_June_24 = pd.read_csv('../data/Inpatient_Rehabilitation_Facility-Provider_Data_Jun2024.csv')
IRF_provider_data_June_24.head(1)
```

```python
NC_Readmissions_2024=IRF_provider_data_June_24.loc[(IRF_provider_data_June_24['Measure Code'] == 'I_018_01_PPR_WI_RSRR')& (IRF_provider_data_June_24['State']=='NC')].reset_index(drop=True)
```

```python
drop_rows= NC_Readmissions_2024.loc[NC_Readmissions_2024["Score"]== "Not Available"]
```

```python
NC_Readmissions_2024=NC_Readmissions_2024.drop(drop_rows.index, axis=0)
```

```python
NC_Readmissions_2024['Score']=NC_Readmissions_2024['Score'].astype(float)
```

```python
NC_Readmissions_2024=NC_Readmissions_2024[['CMS Certification Number (CCN)', 'County/Parish','State','Measure Code','Score']].rename(columns={'Score':'2024 Score'})
```

```python
NC_Avg_Readmissions_2024=NC_Readmissions_2024.groupby('County/Parish')['2024 Score'].mean().to_frame().reset_index()
NC_Avg_Readmissions_2024
```

```python
IRF_provider_data_June_25 = pd.read_csv('../data/Inpatient_Rehabilitation_Facility_Provider_Data_Jun2025.csv')
```

```python
NC_Readmissions_2025=IRF_provider_data_June_25.loc[(IRF_provider_data_June_25['Measure Code'] == 'I_018_01_PPR_WI_RSRR')& (IRF_provider_data_June_25['State']=='NC')].reset_index(drop=True)
```

```python
drop_rows= NC_Readmissions_2025.loc[NC_Readmissions_2025["Score"]== "Not Available"]
```

```python
NC_Readmissions_2025=NC_Readmissions_2025.drop(drop_rows.index, axis=0)
```

```python
NC_Readmissions_2025['Score']=NC_Readmissions_2025['Score'].astype(float)
```

```python
NC_Readmissions_2025=NC_Readmissions_2025[['CMS Certification Number (CCN)', 'County/Parish','State','Measure Code','Score']].rename(columns={'Score':'2025 Score'})
```

```python
Avg_NC_Readmissions_2025=NC_Readmissions_2025.groupby('County/Parish')['2025 Score'].mean().to_frame().reset_index()
Avg_NC_Readmissions_2025
```

Merging all the frames together:

```python
dfs=[Avg_NC_Readmissions_2023, NC_Avg_Readmissions_2024, Avg_NC_Readmissions_2025]

Avg_NC_Readmits_Merged = dfs[0]

for i in range(1, len(dfs)):
    Avg_NC_Readmits_Merged= pd.merge(Avg_NC_Readmits_Merged, dfs[i], on = 'County/Parish', how ='inner')


Avg_NC_Readmits_Merged
```

```python
Avg_NC_Readmits_Merged['County Overall Avg Readmit Rate']= Avg_NC_Readmits_Merged[['2023 Score','2024 Score','2025 Score']].mean(axis=1)
Avg_NC_Readmits_Merged
```

```python
Avg_NC_Readmits_Merged.columns
```

```python
Avg_NC_Readmits_Merged=Avg_NC_Readmits_Merged.sort_values(by='County Overall Avg Readmit Rate', ascending= False)
```

```python
Avg_NC_Readmits_Merged.info()
```

```python
percent_change=[]

for index, row in Avg_NC_Readmits_Merged.iterrows():
    start= row['2023 Score']
    end= row['2025 Score']
    pct_change=((end-start)/start) *100
    percent_change.append(pct_change)

```

```python
Avg_NC_Readmits_Merged['Pct Change']=percent_change
Avg_NC_Readmits_Merged
```

```python
Avg_NC_Readmits_Merged.head(3)
```

```python
Avg_NC_Readmits_Merged.tail(3)
```

Save to CSV:

```python
Avg_NC_Readmits_Merged.to_csv('Avg_NC_Readmits_Merged.csv', index=False)
```

```python
NC_Overall_Readmit_Avg= Avg_NC_Readmits_Merged[['2023 Score','2024 Score', '2025 Score']].mean()
NC_Overall_Readmit_Avg
```

```python
NC_Overall_Readmit_Avg=NC_Overall_Readmit_Avg.to_frame()
NC_Overall_Readmit_Avg
```

```python
NC_Overall_Readmit_Avg=NC_Overall_Readmit_Avg.transpose()
NC_Overall_Readmit_Avg
```
