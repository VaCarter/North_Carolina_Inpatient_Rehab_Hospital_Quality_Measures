Now looking at conditions that were most prevalent in each county across all three years:

```python
dfs=[highest_percent_conditions_county23, highest_percent_condition_county24, highest_conditions_county25]

Combined_highest_pct_condition = dfs[0]

for i in range(1, len(dfs)):
    Combined_highest_pct_condition= pd.merge(Combined_highest_pct_condition, dfs[i], on = 'County/Parish', how ='inner')


Combined_highest_pct_condition
```

```python
Combined_highest_pct_condition=Combined_highest_pct_condition[['County/Parish','Condition_x','Percent_Condition_x','Condition_y','Percent_Condition_y','Condition','Percent_Condition 2025']]
Combined_highest_pct_condition
```

FInding which condition had the highest % by county across the years

```python
max_conditions=[]

for _, row in Combined_highest_pct_condition.iterrows():
    # Get the three percent values
    p_2023 = row['Percent_Condition_x']
    p_2024 = row['Percent_Condition_y']
    p_2025 = row['Percent_Condition 2025']

    percents = {
        '2023': p_2023,
        '2024': p_2024,
        '2025': p_2025
    }

    max_year = max(percents, key=percents.get)
    
    # Get the corresponding condition
    if max_year == '2023':
        max_condition = row['Condition_x']
    elif max_year == '2024':
        max_condition = row['Condition_y']
    elif max_year =='2025':
        max_condition = row['Condition']
    
    # Append to list
    max_conditions.append(max_condition)

```

```python
Combined_highest_pct_condition['Most Common Condition']=max_conditions
Combined_highest_pct_condition
```
