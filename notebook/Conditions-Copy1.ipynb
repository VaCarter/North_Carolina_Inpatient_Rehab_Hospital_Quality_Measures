```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## 2025

```python
Conditions_25=pd.read_csv('../data/Inpatient_Rehabilitation_Facility-Conditions_Jun2025.csv')
```

```python
Conditions_25= Conditions_25[['CMS Certification Number (CCN)','CMS Region','Condition','Count']]
```

```python
rows_to_drop= Conditions_25.loc[(Conditions_25['Count']=='Not Available') | (Conditions_25['Count']=='Less than 11')]
```

```python
Conditions_25=Conditions_25.drop(rows_to_drop.index, axis=0)
Conditions_25
```

```python
Conditions_25.info()
```

```python
Conditions_25['Count']=Conditions_25['Count'].astype(str).astype(float)
```

```python
Conditions_grouped_25 = Conditions_25.groupby(['CMS Region', 'Condition'])['Count'].sum().to_frame()
```

```python
Conditions_grouped_25=Conditions_grouped_25.reset_index()
Conditions_grouped_25
```

```python
Conditions_grouped_25['Percent_Condition'] = (
    Conditions_grouped_25.groupby('CMS Region')['Count']
    .transform(lambda x: x / x.sum() * 100)
)
```

```python
Conditions_grouped_25
```

```python
min_condition=Conditions_grouped_25['Percent_Condition'].min()
min_condition
```

```python
Conditions_grouped_25.loc[Conditions_grouped_25['Percent_Condition']==0.14775898867181086]
```

```python
max_condition=Conditions_grouped_25['Percent_Condition'].max()
max_condition
```

```python
Conditions_grouped_25.loc[Conditions_grouped_25['Percent_Condition']==43.279852709100474]
```
