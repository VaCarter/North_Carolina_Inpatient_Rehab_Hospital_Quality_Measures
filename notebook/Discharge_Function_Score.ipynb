```python
import pandas as pd
```

```python
Provider_Data_25= pd.read_csv('../data/Inpatient_Rehabilitation_Facility_Provider_Data_Jun2025.csv')
```

```python
Provider_Data_24=pd.read_csv('../data/Inpatient_Rehabilitation_Facility-Provider_Data_Sep2024.csv')

# Had to pull from September's quarterly data was not available until Quarter 3.
```

```python
Provider_Data_23=pd.read_csv('../data/Inpatient_Rehabilitation_Facility-Provider_Data_Jun2023.csv')
# there is no data for 2023.  This measure was introduced in 2023, with reporting starting in Quarter 3 of 2024.
```

Filtering down to NC and discharge function measure

```python
Provider_Data_25.head()
```

```python
DC_Function_25_NC= Provider_Data_25.loc[(Provider_Data_25['State']=='NC') & (Provider_Data_25['Measure Code']=='I_026_01_OBS_RATE')].drop(columns=['Address Line 1','Address Line 2', 'City/Town','ZIP Code','Telephone Number','CMS Region','Footnote','Start Date','End Date', 'Measure Date Range']).reset_index(drop=True)
DC_Function_25_NC.head()
```

```python
DC_Function_25_NC=DC_Function_25_NC.drop(columns=['CMS Certification Number (CCN)','Provider Name','State'])
```

```python
DC_Function_25_NC=DC_Function_25_NC.rename(columns={'County/Parish':'County','Score':'2025 Score'})
DC_Function_25_NC.head()
```

```python
rows_to_drop=[]

for index, row in DC_Function_25_NC.iterrows():
    if row['2025 Score']=='Not Available':
        rows_to_drop.append(index)

DC_Function_25_NC.drop(rows_to_drop, inplace=True)
```

```python
DC_Function_25_NC
```

```python
Provider_Data_24.head()
```

```python
DC_Function_24_NC=Provider_Data_24.loc[Provider_Data_24['State']=='NC']
```

```python
DC_Function_24_NC=DC_Function_24_NC.loc[DC_Function_24_NC['Measure Code']=='I_026_01_OBS_RATE']
```

```python
DC_Function_24_NC=DC_Function_24_NC[['County/Parish','Score']]
DC_Function_24_NC.head()
```

```python
DC_Function_24_NC=DC_Function_24_NC.rename(columns={'County/Parish':'County','Score':'2024 Score'}).reset_index(drop=True)
```

```python
rows_to_drop=[]

for index, row in DC_Function_24_NC.iterrows():
    if row['2024 Score']=='Not Available':
        rows_to_drop.append(index)

DC_Function_24_NC.drop(rows_to_drop, inplace=True)
```

```python
DC_Function_24_NC
```

```python
Combined_DC_Function_Score_NC= pd.merge(DC_Function_24_NC, DC_Function_25_NC,
                                        on='County',
                                        how='inner')
Combined_DC_Function_Score_NC
```

```python
Combined_DC_Function_Score_NC.info()
```

```python
Combined_DC_Function_Score_NC['2024 Score']=Combined_DC_Function_Score_NC['2024 Score'].astype(str).astype(float)
```

```python
Combined_DC_Function_Score_NC['2025 Score']=Combined_DC_Function_Score_NC['2025 Score'].astype(str).astype(float)
```

```python
Combined_DC_Function_Score_NC.info()
```

```python
Combined_AVG_DC_Function_Score_NC=Combined_DC_Function_Score_NC.groupby('County').agg({'2024 Score':'mean',
                                                                                   '2025 Score':'mean'})
```

```python
Combined_AVG_DC_Function_Score_NC
```

```python
Combined_AVG_DC_Function_Score_NC['Pct change']=((Combined_AVG_DC_Function_Score_NC['2025 Score']-Combined_AVG_DC_Function_Score_NC['2024 Score'])/(Combined_AVG_DC_Function_Score_NC['2024 Score'])) * 100
Combined_AVG_DC_Function_Score_NC
```

```python
Combined_AVG_DC_Function_Score_NC=Combined_AVG_DC_Function_Score_NC.reset_index()
Combined_AVG_DC_Function_Score_NC
```

```python
Combined_AVG_DC_Function_Score_NC.to_csv('Combined_AVG_DC_Function_Score_NC.csv', index=False)
```
