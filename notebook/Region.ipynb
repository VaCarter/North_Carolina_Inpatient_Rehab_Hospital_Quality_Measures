```python
region_NC_23=pop_projections_2023[['County','Region','Age 65 Plus','Total','Year']]
region_NC_23
```

```python
NC_IRF_Counties_List=['Buncombe','Catawba','Craven','Cumberland','Durham','Forsyth','Guilford','Lenoir','Mecklenburg','Moore','Nash','New_Hanover','Orange','Pitt','Rowan','Vance','Wake']
```

```python
region_NC_23= region_NC_23[region_NC_23['County'].isin(NC_IRF_Counties_List)]
region_NC_23
```

#### Finding the percent of the population that is 65+ by region, and percent growth of 65+ by region

```python
region_NC_23_totalpop=region_NC_23.groupby('Region')['Total'].sum()
region_NC_23_totalpop
```

```python
region_NC_23_totalpop=region_NC_23_totalpop.to_frame()
region_NC_23_totalpop
```

```python
region_NC_23_pct65= region_NC_23.groupby('Region')['Age 65 Plus'].sum().to_frame()
region_NC_23_pct65
```

```python
region_NC_23_pct65=region_NC_23_pct65.rename(columns={'Age 65 Plus':'Age 65 Plus 2023'})
region_NC_23_pct65
```

```python
pct_65plus_by_region_23=pd.merge(region_NC_23_pct65, region_NC_23_totalpop,
                      on='Region',
                      how='inner')
pct_65plus_by_region_23
```

```python
pct_65plus_by_region_23['pct 65 plus']=(pct_65plus_by_region_23['Age 65 Plus 2023'])/(pct_65plus_by_region_23['Total']) * 100
pct_65plus_by_region_23
```

```python
pct_65plus_by_region_23.to_csv('pct_65plus_by_region_23.csv', index=True)
```

#### Finding the Percent growth of age 65 plus from '23-'25 by region

```python
pop_projections_2023.head(1)
```

```python
region_grouped_23=pop_projections_2023[['County','Region','Year','Race','Age 65 Plus', 'Total']]
```

```python
region_grouped_23
```

```python
dfs=[region_grouped_23, region_grouped_24, region_grouped_25]

regions_grouped_merged = dfs[0]

for i in range(1, len(dfs)):
    regions_grouped_merged= pd.merge(regions_grouped_merged, dfs[i], on = 'County', how ='inner')


regions_grouped_merged
```

```python
regions_grouped_merged=regions_grouped_merged.rename(columns={'Age 65 Plus_x':'Age 65 Plus 2023', 'Age 65 Plus_y':'Age 65 Plus 2024','Age 65 Plus':'Age 65 Plus 2025'})
regions_grouped_merged.head()
```

```python
regions_grouped_merged= regions_grouped_merged[['County','Region', 'Age 65 Plus 2023','Age 65 Plus 2024','Age 65 Plus 2025']]
regions_grouped_merged.head()
```

```python
regions_grouped_merged_two= regions_grouped_merged.groupby('Region').agg({'Age 65 Plus 2023':'sum',
                                                                          'Age 65 Plus 2024':'sum',
                                                                          'Age 65 Plus 2025':'sum'})
regions_grouped_merged_two
```

```python
regions_grouped_merged_two['Pct growth 65 plus']= (regions_grouped_merged_two['Age 65 Plus 2025']-regions_grouped_merged_two['Age 65 Plus 2023'])/(regions_grouped_merged_two['Age 65 Plus 2023']) * 100
```

```python
regions_grouped_merged_two=regions_grouped_merged_two.reset_index()
regions_grouped_merged_two
```

```python
regions_grouped_merged_two.to_csv('regions_grouped_merged_two.csv', index=False)
```
