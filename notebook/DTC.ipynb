```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
IRF_provider_data_June_23 = pd.read_csv('../data/Inpatient_Rehabilitation_Facility-Provider_Data_Jun2023.csv')
IRF_provider_data_June_23.head()
```

```python
IRF_provider_data_June_23=IRF_provider_data_June_23[['CMS Certification Number (CCN)','State','County Name','CMS Region', 'Measure Code','Score']]
IRF_provider_data_June_23.head()
```

```python
NC_DTC_Rate_23=IRF_provider_data_June_23.loc[(IRF_provider_data_June_23['Measure Code']=='I_019_02_DTC_OBS_RATE')& (IRF_provider_data_June_23['State']=='NC')]
NC_DTC_Rate_23.head()
```

```python
not_available=NC_DTC_Rate_23.loc[NC_DTC_Rate_23['Score']=='Not Available']
```

```python
NC_DTC_Rate_23= NC_DTC_Rate_23.drop(not_available.index, axis=0).reset_index(drop=True)
```

```python
NC_DTC_Rate_23['Score']=NC_DTC_Rate_23['Score'].astype(str).astype(float)
```

```python
Avg_NC_DTC_Rate_23=NC_DTC_Rate_23.groupby('County Name')['Score'].mean().to_frame().reset_index()
```

```python
Avg_NC_DTC_Rate_23=Avg_NC_DTC_Rate_23.rename(columns={'Score':'2023_Score','County Name':'County/Parish'})
```

```python
Avg_NC_DTC_Rate_23
```
