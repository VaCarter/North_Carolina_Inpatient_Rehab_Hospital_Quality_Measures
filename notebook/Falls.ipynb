```python
Falls_25= Provider_data_25.loc[(Provider_data_25['Measure Code']=='I_013_01_OBS_RATE') | (Provider_data_25['Measure Code']=='I_013_01_NUMERATOR') | (Provider_data_25['Measure Code']=='I_013_01_DENOMINATOR')]
```

```python
Falls_25
```

```python
NC_Falls_25_Rate= Falls_25.loc[(Falls_25['Measure Code']=='I_013_01_OBS_RATE') &(Falls_25['State']=='NC')].drop(columns=['Address Line 1','Address Line 2', 'City/Town','ZIP Code','Telephone Number','Footnote','Start Date','End Date', 'Measure Date Range']).reset_index(drop=True)
```

```python
NC_Falls_25_Rate.head(10)
```

```python
NC_Falls_25_Rate.info()
```

```python
rows_to_drop= NC_Falls_25_Rate.loc[NC_Falls_25_Rate['Score']=='Not Available']
rows_to_drop
```

```python
NC_Falls_25_Rate=NC_Falls_25_Rate.drop(rows_to_drop.index, axis=0)
NC_Falls_25_Rate
```

```python
NC_Falls_25_Rate['Score']=NC_Falls_25_Rate['Score'].astype(str).astype(float)
NC_Falls_25_Rate
```

```python
National_Falls_Rate= National_Data_25.loc[National_Data_25['Measure Code']=='I_013_01_NATL_OBS_RATE']
```

```python
National_Falls_Rate=National_Falls_Rate[['CMS Certification Number (CCN)','Measure Code','Score']]
National_Falls_Rate
```

```python
Avg_NC_Falls_Rate_25= NC_Falls_25_Rate.groupby('County/Parish')['Score'].mean().to_frame().rename(columns={'Score':'2025'}).reset_index()
```

```python
Avg_NC_Falls_Rate_25
```

```python
Avg_NC_Falls_Rate_25.to_csv('Avg_NC_Falls_Rate_25.csv', index=False)
```

```python
Avg_NC_Falls_Rate_25['2025'].sum()
```

```python
Avg_NC_Falls_Rate_25.plot(kind='bar', x='County/Parish', y='2025', figsize=(12,8))

national_average = 0.2
plt.axhline(y=national_average, color='red', linestyle='--', label=f'National Average: {national_average:.2f}')

plt.title('Average Percent of Patients Who Experience At Least One Fall With Major Injury 2025')
plt.xlabel('County Name')
plt.xticks(rotation=45)

plt.ylabel('Avg. Percent of Patients with a Fall')
          
```
