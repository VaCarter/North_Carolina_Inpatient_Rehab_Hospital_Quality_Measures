```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
race_age_groups_23=pd.read_csv('../data/Race_Age_Groups_23_CSV.csv')
```

```python
NC_IRF_Counties_List=['Buncombe','Catawba','Craven','Cumberland','Durham','Forsyth','Guilford','Lenoir','Mecklenburg','Moore','Nash','New_Hanover','Orange','Pitt','Rowan','Vance','Wake']
```

```python
race_age_groups_23=race_age_groups_23[race_age_groups_23['County'].isin(NC_IRF_Counties_List)]
race_age_groups_23
```

```python
race_age_groups_23_old=race_age_groups_23[['County','Region', 'Race','Total','Age 65 Plus','geom','geo_point_2d']]
```

```python
race_age_groups_23_old=race_age_groups_23_old.rename(columns={'Age 65 Plus':'2023_65_Plus'})
race_age_groups_23_old
```

```python
race_age_groups_23_old.to_csv('race_age_groups_23_old.csv')
```

### 2025

```python
race_age_groups_25=pd.read_csv('../data/Race_Age_Groups_25_CSV.csv')
```

```python
race_age_groups_25=race_age_groups_25[race_age_groups_25['County'].isin(NC_IRF_Counties_List)]
```

merging

```python
dfs=[race_age_groups_23_old, race_age_groups_24_old, race_age_groups_25_old]

Race_Age_Groups_Merged = dfs[0]

for i in range(1, len(dfs)):
    Race_Age_Groups_Merged= pd.merge(Race_Age_Groups_Merged, dfs[i], on = 'County', how ='inner')

Race_Age_Groups_Merged.head()
```

```python
Race_Age_Groups_Merged=Race_Age_Groups_Merged[['County','Region','Race','2023_65_Plus','2024_65_Plus','2025_65_Plus','Total','geom', 'geo_point_2d']]
```

```python
Race_Age_Groups_Merged
```

Percentages of Races by County

```python
Race_Age_Groups_Merged['Growth Rate Pop 65 Plus']= (Race_Age_Groups_Merged['2025_65_Plus']-Race_Age_Groups_Merged['2023_65_Plus'])/Race_Age_Groups_Merged['2023_65_Plus']

```

```python
Race_Age_Groups_Merged
```

```python
Race_Age_Groups_Merged.to_csv('Race_Age_Groups_Merged.csv',index=False)
```

```python
race_age_groups_23.columns
```

```python
race_age_groups_23['pct_65_plus']= race_age_groups_23['Age 65 Plus']/race_age_groups_23['Total']
```

```python
Pct_65plus_race_age_groups_23=race_age_groups_23[['County','Region','Race','pct_65_plus']].reset_index(drop=True)
```

```python
Pct_65plus_race_age_groups_23
```

```python
Pct_65plus_race_age_groups_23.to_csv('Pct_65plus_race_age_groups_23.csv',index=False)
```

```python
sns.set(style="whitegrid")

# Increase figure size for readability
plt.figure(figsize=(16, 8))

# Create a bar plot
sns.barplot(
    data=Pct_65plus_race_age_groups_23,
    x='County',
    y='pct_65_plus',
    hue='Race',
    ci=None
)

# Add labels and title
plt.title('Percent of Population 65+ by County and Race 2023')
plt.xlabel('Percent 65+')
plt.ylabel('County')

# Place legend outside the plot
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Race')

# Show the plot
plt.tight_layout()
plt.show()
```
